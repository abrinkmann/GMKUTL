```python
import pandas as pd
import csv
import numpy as np

#Load Data
dfDBPedia = pd.read_csv('DBPediaAuthorsReduced.csv', sep=',')
dfDnb = pd.read_csv('DnbDataAuthorsReduced.csv', sep=',')
dfVIAF = pd.read_csv('VIAFDataAuthorsReduced.csv', sep=',')
dfWiki = pd.read_csv('WikiDataAuthorsReduced.csv', sep=',')

#Lowercase identifier
dfDBPedia['URI'] = dfDBPedia['URI'].str.lower()
dfDnb['uri'] = dfDnb['uri'].str.lower()
dfVIAF['uri'] = dfVIAF['uri'].str.lower()
dfWiki['uri'] = dfWiki['uri'].str.lower()

#Load True Matches
dfDnbGoldStandard = pd.read_csv('goldstandard_author_DnbDataAuthors.csv', sep=',', header=None)
dfVIAFGoldStandard = pd.read_csv('goldstandard_author_VIAFDataAuthors.csv', sep=',', header=None)
dfWikiGoldStandard = pd.read_csv('goldstandard_author_WikiDataAuthors.csv', sep=',', header=None)
```

```python
dfDBPediaInfo = dfDBPedia.filter(items = ['birthdate', 'deathdate', 'gender', 'label', 'work']).describe()

datatypes = {'birthdate': ['date'], 'deathdate': ['date'], 
             'gender': ['string'], 'label': ['string'], 'work': ['list']}

dfDBPediaProfiling = pd.DataFrame(data= datatypes)
dfDBPediaProfiling = dfDBPediaProfiling.append((dfDBPediaInfo.loc[['count']] / len(dfDBPedia)).astype(np.float).round(3))
dfDBPediaProfiling = dfDBPediaProfiling.append((dfDBPediaInfo.loc[['unique']] / dfDBPediaInfo.loc[['count']].values).astype(np.float).round(3))

dfDBPediaProfiling.index = ['Data Type', 'Density', 'Uniqueness']

dfDBPediaProfiling.to_csv('DBPedia_author_profiling.csv', sep=',', encoding='utf-8', index=True, quotechar='"', quoting=csv.QUOTE_ALL)
#del dfDBPediaProfiling['Data Type']
display(dfDBPediaProfiling)
print(dfDBPediaProfiling.transpose().filter(items=['Density', 'Uniqueness']).to_latex())
dfDBPediaProfilingNew = dfDBPediaProfiling.transpose().filter(items=['Density', 'Uniqueness'])
```

```python
dfDnbInfo = dfDnb.filter(items = ['birthdate', 'deathdate', 'gender', 'label']).describe(include='all')

#display(dfDnbInfo)
datatypes = {'birthdate': ['date'], 'deathdate': ['date'], 'gender': ['string'], 'label': ['string']}

dfDnbProfiling = pd.DataFrame(data= datatypes)
display(dfDnb)
dfDnbProfiling = dfDnbProfiling.append((dfDnbInfo.loc[['count']] / len(dfDnb)).astype(np.float).round(3), sort = False)
dfDnbProfiling = dfDnbProfiling.append((dfDnbInfo.loc[['unique']] / dfDnbInfo.loc[['count']].values).astype(np.float).round(3), sort = False)

dfDnbProfiling.index = ['Data Type', 'Density', 'Uniqueness']
```

```python
from nltk import word_tokenize, ngrams
from nltk.metrics import jaccard_distance
import numpy as np

mapping = {"birthdate": "birthdate", "deathdate": "deathdate", "gender": "gender", "label": "label", "work": "work"}
jaccard_similarity_counter = {"birthdate" : 0, "deathdate": 0, "gender":0, "label": 0, "work": 0}

for index, row in dfDnbGoldStandard.iterrows():
    rowDBPedia = dfDBPedia[dfDBPedia['URI'] == row[0]]
    rowDnb = dfDnb[dfDnb['uri'] == row[1]]
    for key in mapping:
        if type(rowDnb[key].values[0]) not in[np.float64, float] and type(rowDBPedia[mapping[key]].values[0]) not in[np.float64, float]:
            jaccard_similarity_counter[key] += jaccard_distance(set(ngrams(rowDBPedia[mapping[key]].values[0], 2)), set(ngrams(rowDnb[key].values[0], 2)))
    
for key in jaccard_similarity_counter:
    jaccard_similarity_counter[key] = round(jaccard_similarity_counter[key] / len(dfDnbGoldStandard), 3)
 
#display(jaccard_similarity_counter)
dfDnbProfiling = dfDnbProfiling.append(pd.DataFrame(data=jaccard_similarity_counter, index=['Heterogeneity']))
#dfDnbProfiling.to_csv('Dnb_author_profiling.csv', sep=',', encoding='utf-8', index=True, quotechar='"', quoting=csv.QUOTE_ALL)
print(dfDnbProfiling.transpose().filter(items=['Density', 'Uniqueness', 'Heterogeneity']).to_latex())
```
