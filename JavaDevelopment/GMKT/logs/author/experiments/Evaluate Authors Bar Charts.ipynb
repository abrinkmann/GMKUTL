```python
# Evaluation of experiments
import matplotlib.pyplot as plt
import pandas as pd

dfExperiments = pd.read_csv('experiments2018-12-15T13-35-47.946.csv', sep=',', low_memory=False, header=None)
dfExperiments.columns = ['Timestamp', 'Experiment', 'Class', 'TargetTable', 'Precision', 'Recall', 'F-1', 'Blocker', 'BlockingFunction', 'TrainedOn', 'TrainedModel']
display(dfExperiments)
```

```python
#Rearrange DataFrame

def rearrange(df):
    dfAllExperimentsRearranged = pd.DataFrame()
    for index, row in df.iterrows():
        rowPrecision = row
        rowPrecision['ResultValue'] = row['Precision']
        rowPrecision['ResultValueType'] = 'Precision'
        
        dfAllExperimentsRearranged = dfAllExperimentsRearranged.append(rowPrecision)
        
        rowRecall = row
        rowRecall['ResultValue'] = row['Recall']
        rowRecall['ResultValueType'] = 'Recall'
        
        dfAllExperimentsRearranged = dfAllExperimentsRearranged.append(rowPrecision)
    
    return dfAllExperimentsRearranged

dfAllExperimentsRearranged = rearrange(dfExperiments)

dfAllExperimentsRearranged = dfAllExperimentsRearranged[~dfAllExperimentsRearranged['Experiment'].str.contains('abstract')]

display(dfAllExperimentsRearranged)
```

```python
display(dfAllExperimentsRearranged.sort_values(by=['F-1'], ascending=False))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plotResults(df, title):
    sns.set(style="whitegrid")
    fig, ax1 = plt.subplots(figsize=(20,8))
    sns.catplot(x="shortExperiment", y="ResultValue", hue="ResultValueType", data=df, height=6, kind="bar", ax=ax1)
    
    titleAll = 'Target: ' + title
    ax1.set_title(titleAll, fontweight="bold", size=20) # Title
    ax1.set_xlabel("Experiments",fontsize = 20.0)
    ax1.set_ylabel("" ,fontsize = 20.0) # Y label
    ax1.tick_params(direction='out', length=6, width=2, labelsize=20, grid_alpha=0.5)
    ax1.tick_params(axis = 'x', labelrotation=45)
    ax1.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=20)

dfDnB = dfAllExperimentsRearranged[('DnbDataAuthors.csv' == dfExperiments['TargetTable']) & dfExperiments['Experiment'].str.contains('baseline')]
dfDnB = dfDnB.append(dfAllExperimentsRearranged[('DnbDataAuthors.csv' == dfExperiments['TargetTable']) & dfExperiments['Experiment'].str.contains('learned')])

dfDnB['shortExperiment'] = dfDnB['Experiment'].str.replace('_author-', '').str.replace('DnbDataAuthors.csv', '')
plotResults(dfDnB, 'DnB Data Set')
```

```python
dfDnB = dfAllExperimentsRearranged[('VIAFDataAuthors.csv' == dfAllExperimentsRearranged['TargetTable']) & dfAllExperimentsRearranged['Experiment'].str.contains('baseline')]
dfDnB = dfDnB.append(dfAllExperimentsRearranged[('VIAFDataAuthors.csv' == dfAllExperimentsRearranged['TargetTable']) & dfAllExperimentsRearranged['Experiment'].str.contains('learned')])

dfDnB['shortExperiment'] = dfDnB['Experiment'].str.replace('_author-', '').str.replace('VIAFDataAuthors.csv', '')
plotResults(dfDnB, 'Viaf Data Set')
```

```python
dfAll = dfAllExperimentsRearranged[('all' == dfAllExperimentsRearranged['TargetTable']) & dfAllExperimentsRearranged['Experiment'].str.contains('baseline')]
dfAll = dfAll.append(dfAllExperimentsRearranged[('all' == dfAllExperimentsRearranged['TargetTable']) & dfAllExperimentsRearranged['Experiment'].str.contains('learned')])

dfAll['shortExperiment'] = dfAll['Experiment'].str.replace('_author-', '').str.replace('all', '')
plotResults(dfAll, 'All Data Sets')
```

```python
#Rearrange data set
dfLabelBaselineExperiments = dfExperiments[dfExperiments['Experiment'].str.contains('learned')]

dfLabelBaselineExperimentsReduced = pd.DataFrame()

for index, row in dfLabelBaselineExperiments.iterrows():
    dfSubSetExperiments = dfExperiments[(dfExperiments['TargetTable'] == row['TargetTable']) & (dfExperiments['Experiment'].str.contains('transfer_matchingRule'))]
    for index2, expRow in dfSubSetExperiments.iterrows():
        expRow['F-1'] = expRow['F-1'] - row['F-1']
        expRow['Precision'] = expRow['Precision'] - row['Precision']
        expRow['Recall'] = expRow['Recall'] - row['Recall']
        dfLabelBaselineExperimentsReduced  = dfLabelBaselineExperimentsReduced.append(expRow)
    
dfLabelBaselineExperimentsReduced = rearrange(dfLabelBaselineExperimentsReduced)
display(dfLabelBaselineExperimentsReduced)
```

```python
dfAllTrained = dfLabelBaselineExperimentsReduced[('all' == dfLabelBaselineExperimentsReduced['TargetTable'])]
dfAllTrained['shortExperiment'] = dfAllTrained['Experiment'].str.replace('_author-', '').str.replace('all', '').str.replace('_matchingRule', '').str.replace('.csv', '').str.replace('Authors', '')
dfAllTrained = dfAllTrained.reset_index()
dfAllTrained =dfAllTrained.reindex([0, 1, 2, 3, 4, 5, 6, 7,14, 15, 8 ,9, 10, 11, 12, 13])
#display(dfAllTrained)

plotResults(dfAllTrained, 'All Data Sets')
```
