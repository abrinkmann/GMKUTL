```python
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
dfMatchingDnB = pd.read_csv('DnbDataAuthors/matchingRule/2018-11-14_09-43-59-249.csv', sep=',')
display(dfMatchingDnB.describe())
display(dfMatchingDnB.corr(method='pearson', min_periods=1).sort_values(by='IsMatch', ascending=False))
```
