```python
#J48 Attribute importance
import pandas as pd
import numpy as np
import os

directory_in_str = 'C:/Users/Alex/Documents/Uni_Mannheim/07_Masterarbeit/Masterthesis-Coding/JavaDevelopment/GMKT/logs/author/learning'
directory = os.fsencode(directory_in_str)

dict_dim = {'Attribute_SimilarityMeasure_Odds_importance_SimpleLogistic' : 'AttributeSimilarityMeasure',
            'Attribute_Odds_importance_SimpleLogistic' : 'Attribute',
            'Attribute_importance_J48': 'Attribute',
            'AttributeSimilarityMeasure_importance_J48': 'AttributeSimilarityMeasure',
            'FeatureThreshold_importance_J48' : 'FeatureThreshold',
            'FeatureThreshold_Odds_importance_SimpleLogistic': 'FeatureThreshold',
            'Feature_importance_J48': 'Feature',
            'Feature_Odds_importance_SimpleLogistic': 'Feature',
            'SplitThreshold_importance_J48': 'SplitThreshold'}
for key in dict_dim:
    dfMatchingLog = pd.DataFrame()
    for file in os.listdir(directory):
        filename = os.fsdecode(file)
        if not '.' in filename:
            extended_directory_str = directory_in_str + '/' + filename + '/matchingRule'
            extended_directory = os.fsencode(extended_directory_str)
            dim_found = []
            for logFile in os.listdir(extended_directory):
                logFileName = os.fsdecode(logFile)
                if key in logFileName:
                    if key not in dim_found:
                        dim_found.append(key)
                        fullLogFilePath = filename + '/matchingRule/' + logFileName
                        dfNewMatchingLog = pd.read_csv(fullLogFilePath, sep=',', encoding='utf-8')
                        dfMatchingLog = dfMatchingLog.append(dfNewMatchingLog)
    if 'SimpleLogistic' in key:
        dfMatchingLog['logOdds'] = np.log(dfMatchingLog['Odds'])
        del dfMatchingLog['Odds']
    display(key)
    dfMatchingLogMean = dfMatchingLog.groupby(by=dict_dim[key]).mean()
    dfMatchingLogMean.columns = [col + '_mean' for col in dfMatchingLogMean.columns]

    dfMatchingLogMeanStd = dfMatchingLog.groupby(by=dict_dim[key]).agg(np.std, ddof=0)
    dfMatchingLogMeanStd.columns = [col + '_std' for col in dfMatchingLogMeanStd.columns]
    
    dfMatchingLogMean['Topic'] = 'Author'
    csvName = key + '_author.csv' 
    
    if 'SimpleLogistic' in key:
        dfMatchingLogMean = dfMatchingLogMean.merge(dfMatchingLogMeanStd, left_index=True, right_index=True).sort_values(by=['logOdds_mean'],ascending=False)
        dfMatchingLogMean = dfMatchingLogMean.round(3)
        display(dfMatchingLogMean)
        dfMatchingLogMean.to_csv(csvName, sep=',')
        print(dfMatchingLogMean.to_latex())
    else:
        dfMatchingLogMean = dfMatchingLogMean.merge(dfMatchingLogMeanStd, left_index=True, right_index=True).sort_values(by=['GainRatioImportance_mean'],ascending=False)
        dfMatchingLogMean = dfMatchingLogMean.round(3)
        display(dfMatchingLogMean)
        dfMatchingLogMean.to_csv(csvName, sep=',')
        print(dfMatchingLogMean.to_latex())

```
