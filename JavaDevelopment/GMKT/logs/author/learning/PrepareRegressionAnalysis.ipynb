```python
import pandas as pd

dfDnBMatchingLog = pd.read_csv('DnbDataAuthors/matchingRule/Attribute_importance_J48_2018-12-18_19-52-35-529_DnbDataAuthors.csv', sep=',', encoding='utf-8')

#dfDnBMatchingLog = dfDnBMatchingLog[dfDnBMatchingLog['Record2Identifier'].str.contains('viaf.org')]

display(dfDnBMatchingLog)
```

```python

uniqueValues = len(pd.unique(dfDnBMatchingLog[['0-RowComparatorLevenshtein-record1Value', '0-RowComparatorLevenshtein-record2Value']].values.ravel('K')))
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

dfSummary = pd.DataFrame()

dfDnBMatchingLogInfo = dfDnBMatchingLog.describe(include='all').transpose()

for index, row in dfDnBMatchingLogInfo.filter(like='postproccesedSimilarity', axis=0).iterrows():
    summary = {}
    dfDnBMatchingLogInfoSubset = dfDnBMatchingLogInfo.filter(like=index.replace('-postproccesedSimilarity', ''), axis=0)
    sRecord1Value = dfDnBMatchingLogInfoSubset.filter(like='-record1PreprocessedValue', axis=0).iloc[0]
    sRecord2Value = dfDnBMatchingLogInfoSubset.filter(like='-record2PreprocessedValue', axis=0).iloc[0]
        
    summary['density'] = (sRecord1Value['count'] + sRecord2Value['count']) /(2*len(dfDnBMatchingLog))
    summary['uniqueness'] = uniqueValues / (sRecord1Value['count'] + sRecord2Value['count'])
        
    summary['mean'] = row['mean']
    summary['std'] = row['std']
    summary['identifier'] = index.replace('-postproccesedSimilarity', '')
    summary['table'] = 'DnB'
        
    dfSummary = dfSummary.append(pd.DataFrame(summary, index=[0]))
    
display(dfSummary)
```
